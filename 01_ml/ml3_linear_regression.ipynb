# Регрессия

Регрессия — класс задач обучения с учителем, когда по определённому набору признаков объекта нужно предсказать целевую переменную.

Задача регрессии — нахождение зависимостей между определяющими переменными и определяемой переменной, если она является непрерывным числом. Например, определить стоимость дома по его площади. Задача линейной регрессии — нахождение такой зависимости, если она линейная.

### Примеры задач регрессии

#### 1. Прогнозирование стоимости недвижимости

В такой задаче в качестве объектов выступают квартиры, а в качестве признаков — характеристики квартиры (площадь, удаленность от метро, этаж и так далее). Признаки могут быть разные: бинарные (есть лифт или нет), количественные (площадь кухни), категориальные (район, тип дома).

Основные особенности такой задачи:
- стоимость меняется со временем, выборка неоднородна и изменчива;
- признаки разных типов, для многих требуется дополнительная предобработка.

#### 2. Прогнозирование объема продаж

В задаче такого типа в качестве объекта выступает комбинация из дня, торговой точки и товара.

Чаще всего здесь встречаются признаки двух типов: **бинарные** (акционный товар или нет, выходной день или нет, торговая точка отдельная или в ТРЦ и так далее) и **количественные** (объем продаж в предыдущие дни).

Основная проблема в задачах такого типа — очень разреженные данные. Поэтому в модели сложнее оценить её качество.

#### 3. Поиск оптимального места для открытия точки общепита и прогнозирование прибыли через год.

В такой задаче мы рассматриваем разные возможные точки для открытия (например, здание или блок под аренду в ТРЦ), которые являются объектами.  Для каждого объекта мы можем предсказать прибыль на год вперед (или другой срок).

Для такой задачи собирается огромное количество признаков всех типов: харатеристики географического расположения точки (метро, транспортная доступность, средняя цена на недвижимость рядом), демографические данные жителей близлежащей территории, данные об объектах, где могут быть потенциальные посетители (вузы и офисы поблизости).

Проблема решения такой задачи заключается в том, что мы имеем мало объектов (вряд ли для расположения точки рассматривается тысяча альтернатив), но при этом у нас много признаков и есть выбросы.

Эти три примера задач можно решить с помощью регрессии. Объединяющая особенность состоит в том, что мы каждый раз предсказываем числовое значение (прибыль, объем продаж, цену).

Итак, если вам необходимо предсказать числовую переменную, и вы имеете данные для обучения модели (регрессия — задача с учителем), то необходимо для решения задачи выбирать регрессионный анализ.

## 2. Линейная регрессия

В простом виде задача линейной регрессии означает, что для набора пар $(x_i, y_i)$ необходимо найти **зависимость** вида
$y_i = f(w, x_i) + \varepsilon _i$
, при этом $f$ — **линейная функция**.

Предполагается, что случайная **ошибка** $\varepsilon$ распределена нормально с нулевым матожиданием и фиксированной дисперсией $\sigma _{\varepsilon}^{2}$ , которая не зависит от переменных $x$ и $y$.

> В общем виде линейная модель будет выглядеть так: $\vec{y} = X \vec{w}\ +  \vec{\varepsilon}$

- Здесь $\vec{y}$ – определяемая переменная. Если бы мы строили модель, определяющую зависимость стоимости дома от его площади – $\vec{y}$ означал бы стоимости всех домов в выборке. Этот вектор имеет размер $(n*1)$, где $n$ – количество домов.
- $X$ - это матрица всех признаков размера $(n*m)$. $n$ – это строка элемента выборки, а $m$ – столбец признака.
- $\vec{w}$ - вектор весов. То, что нам нужно найти в рамках обучения модели линейной регрессии.
- $\vec{\varepsilon}$, в свою очередь, это случайный шум.
- $\vec{y}, X$ нам известны, $\varepsilon$ – случайная переменная, а $\vec{w}$ – то, что нам нужно найти.

## Дополнительные объяснения

Построение регрессионной модели — это поиск такой функции, которая отражает **зависимость одного признака от других**.

Например, ваше настроение зависит от самочувствия, количества работы и погоды за окном. Значит, мы можем построить регрессионную модель, с помощью которой с той или иной точностью можно предсказать ваше настроение — конечно, если у нас будет информация о вашем самочувствии, количестве работы и погоде в вашем населенном пункте.

Или, к примеру, вы хотите научиться предсказывать, как зависит уровень преступности в том или ином регионе от дохода на душу населения. Тогда построив **регрессионную модель**, мы сможем увидеть зависимость между этими двумя признаками и в дальнейшем предсказывать уровень преступности в зависимости от дохода населения в регионе.

![](data/toronto.png)

На изображении выше можно увидеть карту районов города Торонто, на которой отображен результат работы построенной регрессионной модели. Легенда на карте информирует, что красный цвет соответствует значительной взаимосвязи двух переменных: уровней дохода и преступности (здесь стандартное отклонение больше 2.5). Основываясь на этих данных, можно прийти к выводу, что районы Торонто с высоким уровнем преступности имеют низкий доход на душу населения.

Регрессия помогает:
- **Объяснить значение одной одной переменной через другие:** как уровень преступности в регионе варьируется от дохода людей, проживающих в нем.
- **Предсказывать значения зависимой переменной, зная значения независимых:** доход на душу населения в районе, предсказывать уровень преступности.
- **Оценить вклад каждой независимой переменной:** как на уровень преступности влияют плотность населения, количество учебных заведений, парковых зон, наличие достопримечательностей.

Регрессия может быть простой или множественной. В случае простой регрессии мы предсказываем значение зависимой переменной только через одну независимую переменную. В случае множественной — через несколько.

В общем виде уравнение регрессии выглядит следующим образом: $y=a_1{x_1}+a_2{x_2}+...+a_m{x_m}$

- $Y$ — зависимая переменная. Та переменная, которую мы хотим объяснить через другие.
- $X_{1}, X_{2}, …, X_{m}$ — переменные, через которые мы хотим объяснить зависимую переменную. Их называют независимыми переменными/факторами/регрессорами/предикторами.

Например, давайте предскажем ваш уровень счастья, отталкиваясь от продолжительности сна.

Уравнение будет выглядеть примерно так:
*Счастье* $=b *$ продолжительность_сна $+ b1$

Из такого уравнения мы сможем сделать вывод, что если продолжительность увеличивается на $1$, то счастье увеличивается на $b$. А $b1$ — это тот базовый уровень счастья, который есть, даже если совсем не спать.

Простая регрессия в реальной жизни фактически не используется, ведь предсказать значение признака с помощью чего-то одного никогда не получается. Обычно на любое явление влияют многие факторы.

Вернемся к примеру с уровнем счастья. Допустим, мы предсказываем уровень счастья через количество сна, заработную плату и количество свободного времени. Тогда уравнение регрессии будет выглядеть следующим образом:

*Счастье* $=b1*$ *продолжительность_сна* $+b2*$ зарплата $+b3*$ свободное_время $+b$

## 4. Поиск прямой

Мы имеем уравнение $X\beta = y$, где $X$ — прямоугольная матрица размера $m$ на $n$, причём $m>n$ — число строк матрицы больше количества искомых переменных. Для нахождения оптимальных весов $\beta$ можно **минимизировать** модуль разницы между $y$ и $X\beta$.

При минимизации могут возникнуть проблемы, так как у функции модуля нет производной в нуле, поэтому будем оптимизировать квадрат такой разницы:

$ \left \|y-X\beta \right \|^{2}\rightarrow min_{\beta } $

Решение для такой задачи можно получить в виде уравнения:

$ \beta = (X^{T}\: X)^{-1}\: X^{T}\: y $

Из-за проблем предыдущего метода на практике можно использовать метод **оптимизации**, который мы изучали ранее — **градиентный спуск**. То, что мы оптимизируем, выглядит так:

$ L(X;\beta )=\frac{1}{n}\sum_{1}^{n}(x_i^T\beta -y_i)^{2} $

Теперь посчитаем градиент:

$\begin{equation}\begin{array}{l} L(X ; \beta)=\frac{1}{n} \sum_{1}^{n}\left(x_{i}^{T} \beta-y_{i}\right)^{2}=\frac{1}{n} \sum_{i=1}^{n}\left(\left(x_{i}^{T} \beta\right)^{2}-2\left(x_{i}^{T} \beta\right) y_{i}+y_{i}^{2}\right) \\ \frac{d L}{d \beta_{j}}=\frac{1}{n} \sum_{i=1}^{n}\left(2 x_{i j}^{2} \beta_{j}-2 x_{i j} y_{i}\right)=\frac{1}{n} \sum_{i=1}^{n} 2 x_{i j}\left(x_{i j}\beta_{j}-y_{i}\right) \\ \nabla_{\beta} L=\frac{1}{n} \sum_{i=1}^{n} 2 x_{i}^{T}\left(x_{i j}\beta-y_{i}\right) \end{array}\end{equation}$

Изменим значения параметров на каждом шаге оптимизации в соответствии с формулой градиентного спуска:

$ \begin{equation}\beta^{(k)}=\beta^{(k-1)}-\alpha \cdot \frac{1}{n} \sum_{i=1}^{n} 2 x_{i}^{T}\left(x_{i}\beta^{(k-1)}-y_{i}\right)\end{equation} $

где 

- $\beta^{(k-1)}$ вектор, а $y_i$ — число
- $(\beta^{(k-1)} - y_i)$означает вычитание $y_i$ почленно из всех компонент вектора $\beta^{(k-1)}$

### Задание 4.2

Найдите следующий шаг градиентного спуска. Текущая модель: $y = 2x$.

Обучающая выборка:

|  x|  y|
|---|---|
|1 	|2  |
|2 	|5  |

Темп обучения (learning rate): $1/6$

Ответ будет выглядеть как $y = kx + b$, где $k$ и $b$ вы считаете самостоятельно. Дроби округлите до сотых, разделив точкой.

```python
def next_beta(beta, x, y, rate):
    n = x.shape[0]
    grad = 1/n * sum([
        2*x[i].T * (x[i]*beta - y[i])
        for i in range(n)
    ])
    return beta - rate * grad
```

```python
# create matrix
# 1, 1
# 1, 2
w0 = np.array([[1, 1]]).T
x = np.hstack([
    w0,
    np.array([[1, 2]]).T,
])

y = np.array([2, 5])
beta = np.array([0, 2])

k, b = next_beta(beta, x, y, 1/6)
print(f'y = {k:.2f}x + {b:.2f}')
```

### Коэффициент детерминации

$R^2$ — это **доля дисперсии** зависимой переменной, объясняемая рассматриваемой моделью зависимости. Коэффициент можно рассматривать как универсальную меру зависимости одной случайной величины от множества других.

- $ R^{2}=1-\frac{SS_{res}}{SS_{tot}} $
- $ SS_{res}=\sum_{1}^{n}e_{i}^{2}=\sum_{1}^{n}(y_i-\hat{y}_i)^{2} $
- $ SS_{tot}=\sum_{1}^{n}(y_i-\bar{y})^{2}=n\hat{\sigma}_y^{2} $

Интерпретация

- **Коэффициент** детерминации принимает значения от 0 до 1.
- Для **приемлемых** моделей коэффициент должен быть не меньше 50%.
- Модели с коэффициентом детерминации выше 80 % можно признать достаточно хорошими.
- Коэффициент может быть **отрицательным**, это говорит о крайней неадекватности модели: простое среднее приближает лучше.
- Стоит отметить, что коэффициент детерминации довольно часто используется в научных исследованиях (особенно в социальных науках), однако в практических бизнес-задачах он часто бывает **несостоятелен** для оценки качества модели, и об этом необходимо помнить. Он может принимать очень низкие значения при полностью верной модели. И, наоборот, приближаться к единице при неверной модели. Подробнее об этом можно почитать [здесь](https://data.library.virginia.edu/is-r-squared-useless/).

Если пропуски встречаются в большом количестве только в нескольких переменных, которые не являются для нас важными, можно их удалить.

У нас пропуски по сути есть только в одном признаке — жилой площади. Просто не будем брать её в модель.

Далее всегда есть смысл посмотреть на распределение нашей целевой переменной и понять, какие значения мы будем предсказывать:

У нашего распределения есть проблема — слишком сильный перепад. Много квартир в среднем сегменте, но очень мало дорогих квартир. На практике часто в таких случаях логорифмируют переменную, чтобы уменьшить перепады и сгладить хвост.

Смотрим, у каких пар признаков сильная взаимосвязь:

- **Количество школ и количество детских садов**. Довольно логично, так как в Москве детские сады обычно присоединены к школам. Оставляем школы, т.к. у этого признака больше взаимосвязь с целевой переменной.
- **Расстояние до школ и расстояние до детских садов**. По сути, это естественным образом вытекает из предыдущего пункта. Оставляем расстояние до школ.
- **Расстояние до парка и до метро**. Рассуждая аналогично предыдущим парам, оставляем расстояние до метро.
- **Расстояние до Кремля и до бульварного кольца**. У них коэффициент корреляции равен 1, и их корреляция с целевой переменной одинаковая. Поэтому можем убрать любой признак. Убираем расстояние до Кремля.

Итак, нам надо убрать пять признаков (четыре из-за мультиколлинеарности и жилую площадь, так как там много пропусков), остальные оставляем. Теперь наши данные готовы к построению регрессионной модели.
